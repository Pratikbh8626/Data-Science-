# Descriptive Analytics for Numerical Columns

```python
df1 = df[['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']]
```

```python
for col in df1:
    plt.figure()
    sns.histplot(df[col], kde=True)
    plt.title(f"Histogram of {col}")
    plt.savefig(f"{col}_hist.png")
    plt.show()
```

```python
for col in df1:
    plt.figure()
    sns.boxplot(y=df[col])
    plt.title(f"Boxplot of {col}")
    plt.savefig(f"{col}_box.png")
    plt.show()
```

```python
df2 = df[['Date', 'Day', 'SKU', 'City', 'BU', 'Brand', 'Model']]
```

```python
for col in df2:
    plt.figure(figsize=(10,4))
    sns.countplot(data=df, y=col, order=df[col].value_counts().index)
    plt.title(f"Bar Chart for {col}")
    plt.savefig(f"{col}_bar.png")
    plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
numerical_columns = [
    'Volume',
    'Avg Price',
    'Total Sales Value',
    'Discount Rate (%)',
    'Discount Amount',
    'Net Sales Value'
]

scaler = StandardScaler()
standardized = scaler.fit_transform(df[numerical_columns])
standardized_df = pd.DataFrame(standardized, columns=numerical_columns)

df[numerical_columns].head()
standardized_df.head()
```

```python
categorical_columns = [
    'Date',
    'Day',
    'SKU',
    'City',
    'BU',
    'Brand',
    'Model'
]

df_encoded = pd.get_dummies(df,columns = categorical_columns,dtype=int)
(df_encoded.head(10))
```
